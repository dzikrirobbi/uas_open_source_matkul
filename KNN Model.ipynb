```python
from sklearn.datasets import load_breast_cancer

df = load_breast_cancer(as_frame=True)

data_features_df = df.data
data_features_df
```

```python
data = load_breast_cancer()

X = data.data
y = data.target

print(f'Feature:\n{X}')
print(f'Target:\n{y}')
```

```python
feature_names = data.feature_names
target_names = data.target_names

print(f'Feature Names:\n{feature_names}')
print(f'Target Names:\n{target_names}')
```

```python
import matplotlib.pyplot as plt

X = X[:, :2]
plt.scatter(X[:,0], X[:,1], c=y)
plt.title('Sebaran Mean Radius, Mean Texture dan Tingkat Keganasan Kanker')
plt.xlabel('Mean Radius')
plt.ylabel('Worst Area')

plt.grid(True)
plt.show()
```

```python
X_train = X
y_train = y

print(f'x_train:\n{X_train}')
print(f'y_train:\n{y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(X_train, y_train)
```

```python
import numpy as np

mean_radius = 7.22
mean_texture = 28.28 
X_new = np.array([mean_radius, mean_texture]).reshape(1,-1) 
X_new
```

```python
y_new = model.predict(X_new)
y_new

# 0 = malignant (ganas) dan 1 = benign (jinak)
```
